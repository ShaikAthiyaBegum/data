```python
import pandas as pd             #importing pandas
```

```python
covid= pd.read_csv('worldometer_data.csv')    #reading a csv file
covid
```

```python
covid.head(30)             #return top 30 rows
```

```python
covid.tail(10)    #return bottom n
```

```python
covid.set_index('Country/Region',inplace=True)           #changing the index as one of the column
```

```python
covid.sort_index(inplace=True)           #sorting index
```

```python
covid
```

```python
covid.shape                  #Return a tuple representing the dimensionality of the DataFrame.
```

```python
covid.columns                #prints columns
```

```python
covid.count()               #used to count non-NA cells for each column or row
```

```python
covid.dropna(axis='index',how='all')   #drops rows having all null values
```

```python
covid.isna()    #detect missing values and returns boolean values
```

```python
covid.isnull().sum()    #returns the number of missing values in each column
```

```python
covid.fillna('0',inplace=True)       #replaces null values with 0
covid
```

```python
covid.dtypes                  #returns a Series with the data type of each column
```

```python
covid['TotalCases']=covid['TotalCases'].astype(object)  #changing the datatype
covid.describe()                                        #Generate descriptive statistics
```

```python
covid.dtypes                         
```

```python
covid= covid.astype({'Population': 'int64', 'TotalCases': 'int64', 'TotalDeaths': 'int64'}) #changing datatypes of multiple columns
```

```python
covid.corr()        #pairwise correlation of columns
```

```python
import seaborn as sns
```

```python
sns.heatmap(covid.corr())           # Plot rectangular data as a color-encoded matrix
```

```python
covid['Continent'].value_counts()          #returns object containing counts of unique values
```

```python
covid['Continent'].value_counts(normalize=True)    #returns object containing percentage counts of unique values
```

```python
Asia_regions=covid.groupby(['Continent']) #split the data into groups based on criteria
Asia_regions.get_group('Asia')            #we can select a single group
```
